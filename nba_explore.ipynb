```python
import requests
import json

def dis(home, away):
    home = ''.join(map(str,home))[1:-1]
    away = ''.join(map(str,away))[1:-1]
    API_KEY = ''
    # google API url 
    url = f"https://maps.googleapis.com/maps/api/directions/json?origin={away}&destination={home}&mode=driving&key={API_KEY}"
    r = requests.get(url)
    results = json.loads(r.content)
    #return results
    legs = results.get("routes").pop(0).get("legs")
    return legs[0].get("distance")['text'],legs[0].get("duration")['text']
```

```python
from timezonefinder import TimezoneFinder
# object creation
obj = TimezoneFinder()
#obj.timezone_at(lng=latitude, lat=longitude)

def time(home, away):
    home_lat = home.split()[0][1:-1]
    home_lng = home.split()[1][:-1]
    away_lat = away.split()[0][1:-1]
    away_lng = away.split()[1][:-1]
    home_tz = obj.timezone_at(lng=float(home_lng), lat=float(home_lat))
    away_tz = obj.timezone_at(lng=float(away_lng), lat=float(away_lat))
    return home_tz, away_tz
```

```python
dt['timezone'] = dt.apply(lambda x: time(x['home_lat_lon'], x['away_lat_lon']), axis=1)
```

```python
game_total = pd.merge(result, dt,  how='left', on=['home_lat_lon','away_lat_lon'])
```

```python
from pytz import timezone
def tz_diff(date, tz):
    '''
    Returns the difference in hours between timezone1 and timezone2
    for a given date.
    '''
    tz_home = timezone(tz[0])
    tz_away = timezone(tz[1])
    date = pd.to_datetime(date)
    return abs((tz_home.localize(date) - tz_away.localize(date).astimezone(tz_home))).seconds/3600

```

```python
game_total['time_diff'] = game_total.apply(lambda x: tz_diff(x['game_date'], x['timezone']), axis=1)
```
